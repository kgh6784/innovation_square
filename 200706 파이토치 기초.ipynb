### 여러가지 type들

```python
import torch
```

```python
a = torch.IntTensor([1])
print(type(a), a)
b = torch.FloatTensor([1])
print(type(b), b)
c = torch.DoubleTensor([1])
print(type(c), c)
d = torch.LongTensor([1])
print(type(d), d)
e = torch.BoolTensor([True, False])
print(type(e), e)

# Float, Long, Bool이 많이 쓰인다
```

### In-place Operation

```python
x = torch.FloatTensor([[1, 2], [3, 4]])
```

```python
# mul_ 를 더 많이 사용한다. 메모리 재사용시 이게 낫다
```

```python
print(x.mul(2.))
print(x)
print(x.mul_(2.))
print(x)
```

### Zip

```python
# 여러가지 데이터를 묶어서 사용할 때
```

```python
for x, y in zip([1, 2, 3], [4, 5, 6]):
    print(x, y)
```

```python
for x, y, z in zip([1, 2, 3], [4, 5, 6], [7, 8, 9]):
    print(x, y, z)
```

### Mul vs. Matmul

```python
m1 = torch.FloatTensor([[1, 2], [3, 4]])
m2 = torch.FloatTensor([[1], [2]])
print('Shape of Matrix 1: ', m1.shape) # 2 x 2
print('Shape of Matrix 2: ', m2.shape) # 2 x 1
print(m1.matmul(m2)) # 2 x 1

m1 = torch.FloatTensor([[1, 2], [3, 4]])
m2 = torch.FloatTensor([[1], [2]])
print('Shape of Matrix 1: ', m1.shape) # 2 x 2
print('Shape of Matrix 2: ', m2.shape) # 2 x 1
print(m1 * m2) # 2 x 2
print(m1.mul(m2))
```
